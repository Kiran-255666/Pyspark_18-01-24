```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('Titanic-Tutorial').getOrCreate()
```

```
titanic_df = spark.read.csv(path="dbfs:/titanic.csv",header=True,inferSchema=True)
```

```
#convert dataframe to rdd
titanic_rdd = titanic_df.rdd
```

```
type(titanic_rdd)
```

```
titanic_rdd.collect()
```

```
#filter
filter_rdd = titanic_rdd.filter(lambda row : row[ "age"]is not None and row["age"]>25.0 )
```

```
filter_rdd.collect()
```

```
#Map 
mapped_rdd = titanic_rdd.map(lambda row : (row['who'],row['age']))
```

```
mapped_rdd.collect()
```

```
titanic_rdd.collect()
```

```
#Actions
totalpassengers =titanic_rdd.count()
totalpassengers
```

```
first_5_rows = titanic_rdd.take(5)
first_5_rows
```

```
filter_rdd.count()
```

```
total_age =titanic_rdd.filter(lambda row : row[ "age"]is not None ).map(lambda row: row ['age']).reduce(lambda a,b :a+b)
```

```
average_age = total_age/totalpassengers
average_age
```
